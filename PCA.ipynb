```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
col_names = ['target', 'F1R', 'F1S', 'F2R', 'F2S', 'F3R', 'F3S', 'F4R', 'F4S', 'F5R','F5S','F6R','F6S','F7R','F7S','F8R','F8S','F9R','F9S','F10R',
    'F10S',  'F11R','F11S','F12R','F12S','F13R','F13S','F14R','F14S','F15R','F15S','F16R','F16S','F17R','F17S','F18R','F18S','F19R','F19S',   'F20R',
    'F20S','F21R','F21S','F22R','F22S']
spectf= pd.read_table('https://archive.ics.uci.edu/ml/machine-learning-databases/spect/SPECTF.test',sep=',',names=col_names)
df = spectf
df
```

```python
correlation = df.corr()
correlation
```

```python

f, ax = plt.subplots(figsize=(120, 100))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(correlation, annot=True, cmap=cmap)
```

```python
X = df.iloc[:,1:]
y = df.target
np.unique(y)
```

```python
X.describe()
```

```python
# to standardize the data values into a standard format by removing the mean and scaling to unit variance.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaleX = scaler.fit_transform(X)
scaleX
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=10000)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
import time
start = time.time()
model = LogisticRegression(max_iter=10000)
end = time.time()
t1 = (end - start)
print(t1);
```

```python
from sklearn.decomposition import PCA
pca= PCA(n_components=2)
Xpca= pca.fit_transform(X)
```

```python
# ratio of eigenvalue of a particular principal component (eigenvector) with total eigenvalues.
pca.explained_variance_ratio_
```

```python
PCAX_train, PCAX_test, PCAy_train, PCAy_test = train_test_split(Xpca, y, test_size=0.2, random_state=0)
```

```python
s = time.time()
model = LogisticRegression(max_iter=10000)
e = time.time()
print('time of LogisticRegression on pca::')
t2 = e- s
print(t2)
model.fit(PCAX_train,PCAy_train)
print('model accuracy::')
print(model.score(PCAX_test,PCAy_test))
```
